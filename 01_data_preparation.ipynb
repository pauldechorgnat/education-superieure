```python
import pandas as pd
```

## 01 - Renommage des colonnes

```python
df = pd.read_excel(
    "pop-16ans-dipl6820_v2.xlsx",
    sheet_name="COM_2020",
    skiprows=16,
)

df.head()
```

```python
df = df.dropna()
```

```python
diplomes = {
    "rec0": "sans_diplomes",
    "rec1": "diplome_cep",
    "rec2": "diplome_bepc",
    "rec3": "diplome_cap",
    "rec4": "diplome_bac",
    "rec5": "diplome_1e_cycle",
    "rec6": "diplome_2e_cycle",
}

sexes = {
    "s1": "hommes",
    "s2": "femmes",
}

ages = {
    "rec1": "16",
    "rec2": "25",
}

def rename_column(col):
    # print("sexe: ", col[8:10])
    # print("diplome: ", col[4:8])
    # print("age: ", col[15:19])
    sexe = col[8:10]
    diplome = col[4:8]
    age = col[15:19]
    
    new_col = f"{sexes.get(sexe)}-{ages.get(age)}-{diplomes.get(diplome)}"
    return new_col
```

```python
df = df.rename(columns={c: rename_column(c) for c in df.columns[6:]})
```

```python
df.head()
```

```python
columns = {
    "RR": "code_region",
    "DR": "code_departement",
    "CR": "code_arrondissement",
    "STABLE": "indicateur_stabilite_commune",
    "DR22": "code_commune",
}
df = df.rename(columns=columns)
```

## 02 - Suppression de la distinction d'âge

```python
from itertools import product
for sexe, diplome in product(sexes.values(), diplomes.values()):
    df[f"{sexe}-{diplome}"] = df[f"{sexe}-16-{diplome}"] + df[f"{sexe}-25-{diplome}"]
    df = df.drop([f"{sexe}-16-{diplome}", f"{sexe}-25-{diplome}"], axis=1)
```

```python
df.head()
```

## 03 - Construction de la variable enseignement supérieur

```python
diplomes_to_education = {
    "sans_diplomes": "non_education_superieure",
    "diplome_cep": "non_education_superieure",
    "diplome_bepc": "non_education_superieure",
    "diplome_cap": "non_education_superieure",
    "diplome_bac": "non_education_superieure",
    "diplome_1e_cycle": "education_superieure",
    "diplome_2e_cycle": "education_superieure",
}

education_to_diplomes = {}
for k, v in diplomes_to_education.items():
    education_to_diplomes[v] = education_to_diplomes.get(v, []) + [k]

education_to_diplomes
```

```python
for sexe, education in product(sexes.values(), education_to_diplomes.keys()):
    df[f"{sexe}-{education}"] = df[
        [
            f"{sexe}-{diplome}"
            for diplome in education_to_diplomes[education]
        ]
    ].sum(axis=1)
```

```python
for education in education_to_diplomes.keys():
    df[f"{education}"] = df[
        [
            f"{sexe}-{diplome}"
            for diplome in education_to_diplomes[education]
            for sexe in sexes.values()
        ]
    ].sum(axis=1)
```

## 04 - Construction de la variable population totale

```python
df["total"] = df[[education for education in education_to_diplomes]].sum(axis=1)
```

```python
for sexe in sexes.values():
    df[f"total_{sexe}"] = df[[f"{sexe}-{education}" for education in education_to_diplomes]].sum(axis=1)
```

```python
df.head()
```

## 05 - Agregation par région

```python
df_regions = (
    df[list(df.columns[6:]) + ["code_region"]]
    .groupby(["code_region"])
    .sum(numeric_only=True)
    .reset_index()
)
df_regions.head()
```

## 06 - Agregation par région

```python
df_departements = (
    df[list(df.columns[6:]) + ["code_departement"]]
    .groupby(["code_departement"])
    .sum(numeric_only=True)
    .reset_index()
)
df_departements.head()
```

## 07 - Construction des variables de ratio

```python
for c in df.columns[6:-1]:
    df[f"ratio-{c}"] = df[c] / df["total"]
    df_regions[f"ratio-{c}"] = df_regions[c] / df_regions["total"]
    df_departements[f"ratio-{c}"] = df_departements[c] / df_departements["total"]
```

```python
df.head()
```

```python
df = df.dropna()
```

## 08 - Sauvegarde des données

```python
df.to_parquet("clean_data_communes.parquet", index=False)
df_regions.to_parquet("clean_data_regions.parquet", index=False)
df_departements.to_parquet("clean_data_departements.parquet", index=False)
```
